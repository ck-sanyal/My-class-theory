```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from numpy.random import randn

import warnings
warnings.filterwarnings('ignore')
```

```python
x =randn(1000)
```

```python
x
```

```python
sns.distplot(x,kde=True,hist=True)
plt.show()                                                # normal distribution curve
```

```python
sns.kdeplot(x, shade=False)
plt.show()
```

```python
tips=sns.load_dataset('tips')
tips.tail()
```

```python
x=tips['total_bill']
y=tips['tip']
```

```python
sns.jointplot(x=x,y=y)
plt.show()
```
