```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
sns.regplot(x='total_bill',y= 'tip',data=tips)
plt.show()
```

```python
sns.lmplot(x='total_bill',y= 'tip',data=tips)
plt.show()
```
