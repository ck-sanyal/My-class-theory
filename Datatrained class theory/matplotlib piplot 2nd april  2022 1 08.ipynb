```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris=sns.load_dataset('iris')
df=iris.drop(['species'], axis=1)
df.head()
```

```python
d=df.iloc[0]
d
```

```python
d.plot.pie(figsize=(8,8))
plt.show()
```

```python
d=df.head(3).T            # T means transpose
d
```

```python
d.plot.pie(subplots=True,figsize=(25,20))
plt.show()
```

```python
d.plot.pie(subplots=True,figsize=(30,20), fontsize=20,autopct='%.2f')
plt.show()
```
